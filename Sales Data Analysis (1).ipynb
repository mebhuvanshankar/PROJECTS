```python
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df.dropna(subset=['Order Date', 'Sales', 'Product Name', 'Product ID'], inplace=True)
```

```python
df.head()
```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
```

```python
df.head()
```

```python
total_sales = df['Sales'].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
df['Month-Year'] = df['Order Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month-Year')['Sales'].sum().reset_index()
```

```python
monthly_sales['Month-Year'] = monthly_sales['Month-Year'].dt.to_timestamp()
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month-Year', y='Sales', data=monthly_sales)
plt.title('Sales Trends Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
product_sales = df.groupby(['Product ID', 'Product Name'])['Sales'].sum().reset_index()
```

```python
top_products = product_sales.sort_values(by='Sales', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sales', y='Product Name', data=top_products, palette='viridis')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.grid(True)
plt.show()
```
