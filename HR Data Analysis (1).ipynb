```python
import pandas as pd
```

```python
file_path = r"C:\Users\Bhuvanshankar K\Downloads\HR Data.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print("Initial Data:")
print(df.head())
```

```python
print("\nDataFrame Info:")
print(df.info())
```

```python
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
columns_to_remove = ['EmployeeCount', 'Over18', 'StandardHours']  
df.drop(columns=columns_to_remove, inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df.rename(columns={
    'EmployeeNumber': 'Employee_ID',
    'Attrition': 'Employee_Attrition',
    'MonthlyIncome': 'Monthly_Income',
}, inplace=True)
```

```python
df.head()
```

```python
df.drop_duplicates(subset=['Employee_ID'], inplace=True)
```

```python
df.head()
```

```python
string_columns = df.select_dtypes(include=['object']).columns
df[string_columns] = df[string_columns].apply(lambda x: x.str.strip())
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
print("\nMissing Values per Column after cleanup:")
print(df.isnull().sum())
```

```python
df['Employee_Attrition'] = df['Employee_Attrition'].astype('category')
```

```python
print("\nSummary Statistics of the cleaned data:")
print(df.describe(include='all'))
```
