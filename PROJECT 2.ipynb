```python
df = df.drop('Loan_ID',axis = 1)
```

```python
df
```

```python
columns = ['Dependents', 'Gender', 'LoanAmount', 'Loan_Amount_Term']
df = df.dropna(subset = columns)
df.isnull().sum()
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0}).astype('int')
df['Married'] = df['Married'].map({'Yes':1,'No':0}).astype('int')
df['Education'] = df['Education'].map({'Graduate':1,'Not Graduate':0}).astype('int')
df['Self_Employed'] = df['Self_Employed'].map({'Yes':1,'No':0}).astype('int')
df['Property_Area'] = df['Property_Area'].map({'Rural':0,'Semiurban':2,'Urban':1}).astype('int')
df['Loan_Status'] = df['Loan_Status'].map({'Y':1,'N':0}).astype('int')
df.sample(10)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
columns = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x[columns]=st.fit_transform(x[columns])
```

```python
x.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

regressor = LogisticRegression()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)

print(f"{regressor}) accuracy is {accuracy_score(y_test,y_pred)*100}%")

cvscore = cross_val_score(regressor,x,y,cv=5)
print(f"{regressor}) Avg cross val score is {np.mean(cvscore)*100}%")

```

```python
from sklearn.ensemble import RandomForestClassifier

regressor = RandomForestClassifier()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)

print(f"{regressor}) accuracy is {accuracy_score(y_test,y_pred)*100}%")

cvscore = cross_val_score(regressor,x,y,cv=5)
print(f"{regressor}) Avg cross val score is {np.mean(cvscore)*100}%")

```

```python
from sklearn.tree import DecisionTreeClassifier

regressor = DecisionTreeClassifier()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)

print(f"{regressor}) accuracy is {accuracy_score(y_test,y_pred)*100}%")

cvscore = cross_val_score(regressor,x,y,cv=5)
print(f"{regressor}) Avg cross val score is {np.mean(cvscore)*100}%")
```
